```python
# สร้างฟังก์ชั่นเพื่อทำ pivot table ใหม่
def create_user_summary(df):
    users = df['user_id'].unique()
    user_data = []

    for user in users:
        user_df = df[df['user_id'] == user].sort_values(by='datetime')
        user_dict = {'user_id': user}

        for idx, row in enumerate(user_df.itertuples(), start=1):
            user_dict[f'date_{idx}'] = row.datetime
            user_dict[f'package_id_{idx}'] = row.package_id
            user_dict[f'price_{idx}'] = row.price
            user_dict[f'point_{idx}'] = row.point

        user_data.append(user_dict)

    return pd.DataFrame(user_data)

# เรียกใช้ฟังก์ชั่นเพื่อสร้างตารางใหม่
```

```python
# สร้างฟังก์ชั่นเพื่อทำ pivot table ใหม่
def create_user_summary(df):
    users = df['user_id'].unique()
    user_data = []

    for user in users:
        user_df = df[df['user_id'] == user].sort_values(by='datetime')
        user_dict = {'user_id': user}

        for idx, row in enumerate(user_df.itertuples(), start=1):
            user_dict[f'date_{idx}'] = row.datetime
            user_dict[f'package_id_{idx}'] = row.package_id
            user_dict[f'price_{idx}'] = row.price
            user_dict[f'point_{idx}'] = row.point

        user_data.append(user_dict)

    return pd.DataFrame(user_data)

# เรียกใช้ฟังก์ชั่นเพื่อสร้างตารางใหม่
```
