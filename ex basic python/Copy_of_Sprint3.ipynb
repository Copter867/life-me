```python
# แปลงคอลัมน์ datetime เป็น string
df_pay['datetime'] = df_pay['datetime'].astype(str)

# แยกวันที่และเวลาออกจากกัน
df_pay['datetime'] = df_pay['datetime'].str.split(' ')

# เก็บเฉพาะวันที่
df_pay['date'] = df_pay['datetime'].str[0]

# เก็บเฉพาะเวลา
df_pay['time'] = df_pay['datetime'].str[1]

# แปลงวันที่เป็นรูปแบบ datetime
df_pay['date'] = pd.to_datetime(df_pay['date'])

# แปลงเวลาเป็นรูปแบบ datetime
df_pay['time'] = pd.to_datetime(df_pay['time'])

# รวมวันที่และเวลาเข้าด้วยกัน
df_pay['datetime'] = pd.to_datetime(df_pay['date'].dt.strftime('%Y-%m-%d') + ' ' + df_pay['time'].dt.strftime('%H:%M:%S'))

# ลบ columns 'date' และ 'time' ที่ไม่จำเป็น
df_pay.drop(columns=['date', 'time'], inplace=True)
```

```python
# แปลงคอลัมน์ datetime เป็น string
df_pay['actual_time'] = df_pay['actual_time'].astype(str)

# แยกวันที่และเวลาออกจากกัน
df_pay['actual_time'] = df_pay['actual_time'].str.split(' ')

# เก็บเฉพาะวันที่
df_pay['date'] = df_pay['actual_time'].str[0]

# เก็บเฉพาะเวลา
df_pay['time'] = df_pay['actual_time'].str[1]

# แปลงวันที่เป็นรูปแบบ datetime
df_pay['date'] = pd.to_datetime(df_pay['date'])

# แปลงเวลาเป็นรูปแบบ datetime
df_pay['time'] = pd.to_datetime(df_pay['time'])

# ตัดมิลลิวินาทีออกจากเวลา
df_pay['time'] = df_pay['time'].dt.floor('s')

# รวมวันที่และเวลาเข้าด้วยกัน
df_pay['actual_time'] = pd.to_datetime(df_pay['date'].dt.strftime('%Y-%m-%d') + ' ' + df_pay['time'].dt.strftime('%H:%M:%S'))

# ลบ columns 'date' และ 'time' ที่ไม่จำเป็น
df_pay.drop(columns=['date', 'time'], inplace=True)
```

```python
#6-11= morning, 12-17 = afternoon, 18-23 = evening, 0-5 = night
def categorize_data(datetime_value):

  hour = datetime_value.hour
  time_range = 'Morning' if 6 <= hour <= 11 else (
        'Afternoon' if 12 <= hour <= 17 else (
            'Evening' if 18 <= hour <= 23 else 'Night')
    )

  return time_range
```

```python
df_pay2['time_range'] = df_pay2['datetime'].apply(categorize_data)
df_pay2
```
